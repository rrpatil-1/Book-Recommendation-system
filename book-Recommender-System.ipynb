```python
newpopular_df=popular.merge(avgdf,on='Book-Title')
```

```python
rdf=newpopular_df[newpopular_df['Book-Rating']>=300].sort_values(by='avg-rating',ascending=False)
```

```python
r_df=rdf.merge(Books,on='Book-Title').drop_duplicates('Book-Title')
```

```python
x=rwn.groupby('User-ID').count()['Book-Title']>200
exp_user=x[x].index
```

```python
filtered_rating=rwn[rwn['User-ID'].isin(exp_user)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famousbooks=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famousbooks)]
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score[0][0]
```

```python
def recommend(book):
    index=np.where(pt.index==book)[0][0]
    similar_item=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_item:
        item=[]
        temp_df=Books[Books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        print(pt.index[i[0]])
```

```python
recommend('harry potter and the prisoner of azkaban (book 3)')
```

```python
def recommend(book):
    index=np.where(pt.index==book)[0][0]
    similar_item=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_item:
        item=[]
        temp_df=r_df[r_df['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(np.round(temp_df.drop_duplicates('Book-Title')['avg-rating'].values,2)))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```
